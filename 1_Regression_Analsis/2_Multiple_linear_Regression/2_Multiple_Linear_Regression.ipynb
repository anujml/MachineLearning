#### Introduction to Multiple Linear Regression (MLR)

There are two types of regression problems: `simple linear regression` and `multiple linear regression`. In `simple linear regression`, we work with one independent variable and one dependent variable, meaning there is only a single feature used to predict the target variable.

The equation $Y = mX + b$ is used to define the straight line, which has been used to predict the target variable.

In `multiple linear regression`, we work with two or more independent variables, meaning multiple features are used to predict the target variable.

The equation to predict the target variable is:

$$
Y = \beta_0 + \beta_1 \cdot X_1 + \beta_2 \cdot X_2 + \beta_3 \cdot X_3 + \dots + \beta_n \cdot X_n
$$

Where:
- \( Y \) is the target variable.
- \( X \) represents the independent variables.
- \( $\beta_0$ \) is the y-intercept.
- \( $\beta_1, \beta_2 ... \beta_n$ \) are the slopes of the hyperplane.

Formula to Calculate Beta values. Beta values, means all the coefficent or slope and the intercept.

$\beta = (X^T \cdot X)^{-1} \cdot (X^T \cdot Y)$

### Mathematical Intuition of MLR

Using mathematical concepts, we will understand how we arrive at the solution for \( \beta \) as:

$$
\beta = (X^{T} \cdot X)^{-1} \cdot (X^{T} \cdot Y)
$$

We will derive the equation:

$$
Y = \beta_0 + \beta_1 \cdot X_1 + \beta_2 \cdot X_2 + \beta_3 \cdot X_3 + \dots + \beta_n \cdot X_n
$$

with the help of `matrix algebra` and `matrix differentiation`.

$
\textcolor{red}{{Y}^{T}\cdot \hat{Y} = \hat{Y}^{T} \cdot {Y}}

#### Deferentiate the loss function with respect to $\beta$

Now next step is to differentiate the loss function with respect to $\beta$:

before that replace $ \hat{Y} = X \cdot \beta $ mention in equation 1

$ \frac{d}{d \beta} L = (\frac{d}{d \beta})(Y^T Y)  - (\frac{d}{d \beta}) (2 Y^T X \beta ) + (\frac{d}{d \beta})(X \cdot \beta)^{T}(X \cdot \beta) $

$ \frac{d}{d \beta} L = 0 - 2Y^TX + \textcolor{red}{(\frac{d}{d \beta})( \beta^{T} X^{T}X  \beta)} $

lets solve this equation $\textcolor{red}{(\frac{d}{d \beta})( \beta^{T} X^{T}X  \beta)} -- eq.. (1)$

defferentiation of $\beta$ is 1

But there is a $ \beta^T$

so for that we have to use a formula:

where $ a = X^TAX$

$\frac{d}{da} a = 2X^TA$

so in case of : $\textcolor{red}{(\frac{d}{d \beta})( \beta^{T} X^{T}X  \beta)}$

$ X =  \beta$<br>
$ A =  X^{T}X$

so,
$\textcolor{red}{(\frac{d}{d \beta})( \beta^{T} X^{T}X  \beta)} = 2 \beta^T X^TX$

Replace this in equation 1, so output will be:

$ \frac{d}{d \beta} L = -2Y^TX + \textcolor{red}{2 \beta^T X^TX} $

And by using this above formula we can find the slope of $L(\beta)$ with respect to $\beta$

Now we just need to compare it with zero:

$ \frac{d}{d \beta} L = 0 $

$ -2Y^TX + 2 \beta^T X^TX = 0$

$ 2Y^TX = 2 \beta^T X^TX $

$ Y^TX =  \beta^T X^TX $

$ Y^TX =  \beta^T X^TX $

$ Y^TX \cdot (X^TX)^{-1} = \beta^T$

Apply transpose to both side

$ (Y^TX \cdot (X^TX)^{-1}) ^ T = (\beta^T) ^ T$

$ \beta = ((X^TX)^{-1})^T(X^TY)$

Here this $(X^TX)^{-1}$ matrix is a symetric in nature, so transpose of this matrix will return the same matrix, so in that case we can use the same matrix despite using transpose one, so our final out will be

$$ \beta = (X^TX)^{-1}(X^TY)$$

### Implimanting this using code

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = load_diabetes()['data']
y = load_diabetes()['target']
```

```python
class OurLr():
    def __init__(self,intercept = None,slope = None):
        self.intercept = None
        self.slope = None

    def fit(self,X_train,Y_train):
        X_train = np.insert(X_train,0,1,axis = 1)
        X_t = X_train.T
        betas = np.linalg.inv(np.dot(X_t,X_train)).dot(X_t).dot(Y_train)
        self.intercept = betas[0]
        self.slope = betas[1:]
    def predict(self,X_test):
        predicted_value = X_test.dot(self.slope) + self.intercept
        return predicted_value
```

```python
Lr = OurLr()
Lr.fit(X,y)
prediction_Lr = Lr.predict(X)
mse_m1 = mean_squared_error(y,prediction_Lr)
```

```python
Lr2 = LinearRegression()
Lr2.fit(X,y)
prediction_Lr2 = Lr2.predict(X)
mse_m2 = mean_squared_error(y,prediction_Lr2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(prediction_Lr,y,'o');
plt.title(f"Prediction Vs Actual : Selfmade Model ; MSE {round(mse_m1,2)}")

plt.subplot(1,2,2)
plt.plot(prediction_Lr2,y,'o');
plt.title(f"Prediction Vs Actual : SKModel ; MSE {round(mse_m2,2)}")
```
